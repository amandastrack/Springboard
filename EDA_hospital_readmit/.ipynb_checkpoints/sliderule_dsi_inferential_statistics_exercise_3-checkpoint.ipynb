# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

****
## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

#### A. Do you agree with the above analysis and recommendations? Why or why not?
I do not agree with the above analysis because you cannot make these conclusions solely from a scatterplot. We need to perform hypothesis testing on the data to make a correct recommendation.

#### B. Provide support for your arguments and your own recommendations with a statistically sound analysis.
$H_0$: There is no correlation between number of discharges and excess readmission rate.

$H_A$:: There is correlation between the number of discharges and excess readmission rate.

Test Statistic: Pearson correlation coefficient, $ρ$


##### I will use pearsonr from scipy.stats.

```python
from scipy.stats import pearsonr
pearson_corr, p_val = pearsonr(x, y)
print("Pearson Correlation Coefficient:", pearson_corr)
print("P-Value:", p_val)
```

```python
alpha = .01
if p_val < alpha:
    print("p-value is less than our significance level of .01")
else:
    print("p-value is greater than or equal to our significance level of .01")
```

##### Statistical significance vs. Practical significance
Our test proves statistical significance to reject the null hypothesis. Therefore, we can claim that there is a correlation between the number of discharges and excess readmission rate. The Pearson Correlation Coefficient of -0.09 tells us that as the number of discharges increases, the rate of excess readmissions decreases. 

However, the correlation coefficient is very weak and we cannot conclude that there is a strong correlation between number of discharges and excess readmission rates. Therefore, even though there is statistical significance, there is no practical significance. 

Since there is no practical significance, I would not recommend that smaller hospitals should be required to demonstrate upgraded resource allocation for quality care to continue operation. 


##### Scatterplot Disadvantages
In the scatterplot above, it is difficult to see the correlation between number of discharges and rate of excess readmissions. A more imformative plot would include a regression line to clearly see the relationship.

```python
sns.set(style="darkgrid", color_codes=True)
sns.jointplot('Number of Discharges', 'Excess Readmission Ratio', data=clean_hospital_read_df,
              kind='reg', color='r', height=7)
plt.show()
```
